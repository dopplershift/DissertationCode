```python
import numpy as np
import quantities as pq
%matplotlib inline
```

```python
from disser.experiments import load_model_experiments
```

```python
data_cache = load_model_experiments('ref_runs/', 'Cband*')
```

```python
dat = data_cache['C', 'Combined']
```

```python
from disser import datatypes
dat.fields.default_keys['source'] = 'average'
snr = dat.fields.grabData(datatypes.SNR)
phi = dat.fields.grabData(datatypes.PhiDP)
z = dat.fields.grabData(datatypes.Reflectivity)
truea = dat.fields.grabData(datatypes.Attenuation, source='calc')
```

```python
from disser.atten import attenAlgs, algorithms
attenAlgs.runAll(dat, var='H')
from disser.atten.calc import calc_specific_atten
calc_specific_atten(dat)
```

```python
#tuned_zphi.gamma_cache = list()
b,gamma = algorithms.zphi_coeffs['C', 'H']
sc = algorithms.self_consistent(snr, z, phi, b, gamma, dat.gate_length, dat.phi_offset)
sc_spec = np.gradient(sc, 1, dat.gate_length)[1].rescale('dB/km')
msc = algorithms.modified_self_consistent(snr, z, phi, b, gamma, dat.gate_length, dat.phi_offset)
msc_spec = np.gradient(msc, 1, dat.gate_length)[1].rescale('dB/km')
truea_spec = np.gradient(truea, 1, dat.gate_length)[1].rescale('dB/km')
```

```python
g = list()
for ray in range(phi.shape[0]):
    mask = (snr[ray] > 20) & (~np.isnan(phi[ray]))
    if not np.any(mask):
        g.append(np.nan)
    else:
        g.append(truea[ray, mask][-1] / phi[ray, mask][-1])
g = np.array(g)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(12, 8), dpi=200)
az = dat.az[:] * 1
az[az>180] = az[az>180] - 360 * pq.deg
plt.plot(az, algorithms.modified_self_consistent.gammas, label='Raw')
#plt.plot(az, modified_self_consistent.org_gammas, label='Raw')
plt.plot(az, g, label='True')
plt.axhline(algorithms.modified_self_consistent.opt_gamma, color='k', linestyle='--', label='Median')
plt.legend(loc='upper right')
plt.xlim(-45, 45)
plt.xlabel('Azimuth')
plt.ylabel('Best $\gamma$')
plt.ylim(0, 0.25)
```

```python
fig,(ax1,ax2) = plt.subplots(1, 2, figsize=(16, 8), dpi=200, sharex=True, sharey=True)
ax1.plot(truea_spec.flatten(), sc_spec.flatten(), 'o')
ax1.plot([0, 5], [0, 5], 'r--')
ax1.set_title('Self-Consistent')
ax1.set_xlabel('True $A_H$')
ax1.set_ylabel('Algorithm $A_H$')
ax2.plot(truea_spec.flatten(), msc_spec.flatten(), 'o')
ax2.plot([0, 5], [0, 5], 'r--')
ax2.set_title('Modified Self-Consistent')
ax2.set_xlabel('True $A_H$')
ax2.set_ylabel('Algorithm $A_H$')
plt.ylim(0, 3)
plt.xlim(0, 3);
```
