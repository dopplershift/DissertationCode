```python
import numpy as np
import scipy.integrate as si
import matplotlib.pyplot as plt
```

```python
%pylab inline
```

```python
# Z needs to be linear here
def I(z, dr, b):
    return 0.46 * b * si.cumtrapz(z**b, dx=dr, initial=0)
```

```python
# And here too
def I0(z, dr, b):
    return 0.46 * b * np.trapz(z**b, dx=dr)
```

```python
def zphi(z, dr, delta_phi, b=0.03, gamma=0.01):
    z_lin = 10 ** (z / 10.)
    phi_factor = 10 ** (0.1 * b * gamma * delta_phi) - 1
    return phi_factor * z_lin**b / (I0(z_lin, dr, b) + phi_factor * I(z_lin, dr, b))
```

```python
N0 = 8e6
a = 3.64e-6
b = 0.7644
alpha = 1.97
beta = 0.973
```

```python
z = np.arange(0, 80)
z_lin = 10**(z/10.)
A = a * N0 ** (1 - b) * z_lin ** b
```

```python
plt.plot(np.log10(z_lin / N0), np.log10(A / N0), 'k--')
plt.xlim(-7, 1)
plt.ylim(-11, -5)
```

```python
gamma = 3.2e-1
z_diffs = np.random.randn(100)
z = 40 + z_diffs.cumsum()
#z = np.empty(100)
#z.fill(40)
delta_r = 1
z_lin = 10**(z/10.)
A = a * N0 ** (1 - b) * z_lin ** b
atten = (2 * A * delta_r).cumsum()
kdp = alpha * N0 ** (1 - beta) * A ** beta
kdp = A / gamma
phidp = 2 * np.trapz(kdp, dx=delta_r)
```

```python
rng = delta_r * np.arange(z.size)
atten_z = z - atten
corr_z = atten_z + (2 * delta_r * zphi(atten_z, delta_r, phidp, b, gamma)).cumsum()
plt.plot(rng, z, 'b', label='truth')
plt.plot(rng, atten_z, 'r', label='attenuated')
plt.plot(rng, corr_z, 'g', label='corrected')
```
