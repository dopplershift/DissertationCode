```python
# Some initial module imports
import sympy
import sympy.abc
%load_ext sympyprinting
%pylab inline
```

```python
# Set up some symbols for symbolic math
from sympy import gamma, exp, oo, pi
rho_l, rho, q = sympy.symbols('rho_l rho q')
lam, mu, mu_v = sympy.symbols('Lambda mu mu_v', positive=True, imaginary=False)
N0 = sympy.Symbol('N0')
```

## Theoretical Calculation

First, we can calculate the moments of our modified Gamma distribution of drop diameter using the usual relationship:

$$
\int_{0}^{\infty} N_0 D^\mu e^{-\Lambda D} dD = N_0 \Gamma(\mu + 1) \Lambda^{-(\mu + 1)}
$$

We'll use here the zeroth, third, and sixth moments. The zeroth and third are proportional to the number concentration and liquid water content, respectively. The sixth moment is chosen both because it is proportional to the Rayleigh reflectivity factor and because it allows for some convenient cancellation, as well see. We can create the expressions for these three moments below:

```python
D0 = N0 * gamma(mu + 1) * lam ** -(mu + 1)
D3 = N0 * gamma(mu + 4) * lam ** -(mu + 4)
D6 = N0 * gamma(mu + 7) * lam ** -(mu + 7)
display(D0, D3, D6)
```

Creating the proper ratio, we see most terms cancel out and we're left with a product of a few terms as a result of the expansion of the $\Gamma$ function.

```python
F_left = D3**2 / (D0 * D6)
F_left.simplify()
```

We can also create a similar expansion of moments for the distribution of drops in terms of volume. In this case, we use the zeroth, first, and second moments. These are proportional to the same physical quantities as in the distribution in terms of drop diameter.

```python
V0 = N0 * gamma(mu_v + 1) * lam ** -(mu_v + 1)
V1 = N0 * gamma(mu_v + 2) * lam ** -(mu_v + 2)
V2 = N0 * gamma(mu_v + 3) * lam ** -(mu_v + 3)
display(V0, V1, V2)
```

So we set the physical variables $N_r$, $q_r$, and $Z_r$, which correspond to rain number concentration, liquid water content, and reflectivity factor, respectively, to the proper expression in terms of moments of the volume distribution:

```python
Nr = V0
qr = V1 * rho_l
Zr = 36 * V2 / pi**2
```

and now set the diameter distribution moments from these same values:

```python
D0 = Nr
D3 = (6 / (pi * rho_l)) * qr
D6 = Zr
```

If we take the same ratio as above, we see:

```python
F_right = D3**2 / (D0 * D6)
F_right.simplify()
```

When attempting to retrieve a constrained modified Gamma distribution from these 3 moments, this corresponds to the left side of the equation in $\mu$ that is solved above. This implies that a fixed shape in the volume distribution yields a fixed shape in the diameter distribution (when using these 3 moments as constraints). One benefit of this fact is that given this volume shape factor, the use of the sixth moment of the distributions is superfluous. We can substitute our fixed value of -0.8 into the equation and solve the resulting polynomial:

```python
result = F_right.simplify().evalf(subs={mu_v:-0.8})
result
```

```python
sympy.solve(sympy.Eq(F_left.simplify(), result))
```

Thus, taking the only physical solution, we have:

$$
\mu = 1.81028329387715
$$

## Impact of changing distribution

Let's see what difference this makes in the actual shape of the distribution, using a representative sample of $q_r = 4 g m^{-3}$ and $N_r = 8000 m^{-3}$.

```python
# Some more imports
import numpy as np
import matplotlib.pyplot as plt
import quantities as pq
import dsd
plt.rcParams['savefig.dpi'] = 108
figsize(9,6)
```

```python
qr = 4 * pq.g / pq.m**3
nr = 8000 / pq.m**3
d = np.linspace(0.0001, 8, 10000) * pq.mm
```

```python
final_units = 'mm^-1 * m^-3'
# First calculate the old volume gamma distribution
volume_gamma = dsd.volume_gamma(d, dsd.gamma_d0_from_lwc(qr, nr), nr, nu=-0.8).rescale(final_units)

# Now the modified gamma
gamma_shape = 1.81028329387715
gamma_slope = dsd.gamma_slope(nr, qr, gamma_shape)
gamma_intercept = dsd.gamma_intercept(nr, gamma_slope, gamma_shape)
modified_gamma = dsd.modified_gamma(d, gamma_slope, gamma_intercept, gamma_shape).rescale(final_units)

# And now exponential and marshall-palmer for reference
exponential = dsd.exponential_from_moments(d, nr, qr).rescale(final_units)
marshall_palmer = dsd.mp_from_lwc(d, qr).rescale(final_units)
```

We can plot these distributions together to compare their appearance. It's clear that moving to the modified gamma has produced a distribution with a much more expected shape to it than the volume gamma, but also better than just simple exponential or Marshall-Palmer.

```python
plt.semilogy(d, volume_gamma, 'k-', label='Volume Gamma')
plt.semilogy(d, modified_gamma, 'b', linewidth=2, label='Modified Gamma')
plt.semilogy(d, exponential, 'k--', label='Exponential')
plt.semilogy(d, marshall_palmer, 'k:', label='Marshall-Palmer')
plt.ylim(1, None)
plt.xlabel('Diameter (%s)' % d.dimensionality.latex)
plt.ylabel('Distribution (%s)' % marshall_palmer.dimensionality.latex)
plt.legend()
plt.grid()
```

For reference, we can integrate each of these distributions to check that they still give the correct parameters. All are correct, other than the number concentration for Marshall-Palmer, which cannot match the true number concentration due to the fixed intercept parameter.

```python
# Loop over all the distributions and calculate some integral moments
for dist in [volume_gamma, modified_gamma, exponential, marshall_palmer]:
    print 'qr: %s Nr: %s' % (dsd.lwc(d, dist).simplified, np.trapz(dist, x=d).simplified)
```
